# 0.  Load the darplotter library

```python
# import darkplotter lib
import darkplotter as dp
```

# 1. Call the Data manager constructor

```python
mydata = dp.DMdata()
```

# 2. Upload all the experiments
## A folder with experiement json can be loaded
by default this folder
https://github.com/odadoun/DarkPlotter/tree/NewGraphicToAdd/json/
A conversion between github file path and rawgithub need to be done 

```python
#Url of the json folder
u=mydata.githubpath2raw()
print(len(u))
mydata.uploadexperiement(url=u.rawurl.tolist())
```

# 3. Select the experiment you want
## By setting "collaboration" and "experiment", default plot includes all the experiment

```python
#Select experiments by setting "collaboration" and "experiment", no parameter will plot all the experiments
#myexp=mydata.get_experiment()
myexp=mydata.get_experiment(collaboration=["XENON","LUX","DAMA","neutrino","CDMS"],experiment=["LUX","XENON","Coherent Neutrino Scattering","CDMS II","XENON 100 S2"])
myexp.set_index("experiment")[:5]
```

# 4. Call the DMplotter constructor 
## Change the massunit by setting "massunit"

```python
myplotter = dp.DMplotter()
myplotter.plot(myexp,massunit="GeV")
```
